```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [5,3,5,6,2,3,5,6,9]


plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [5,3,5,6,2,3,5,6,9]


plt.plot(x,y, marker="*",color='green')
plt.title('test')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = ['red','green','blue','white']
y = [5,19,4,14]

plt.bar(x,y)
plt.show()
```

```python
x = ['red','green','blue','white']
y = [5,19,4,14]

plt.bar(x,y , color='purple')
plt.title('test')
plt.xlabel('color')
plt.ylabel('count')
plt.show()
```

```python
x = ['red','green','blue','pink']
y = [5,19,4,14]

plt.pie(y, labels=x, autopct='%1.1f%%', colors=['red','green','blue','pink'])
plt.show()
```

```python
import numpy as np

x = np.random.rand(1000, 1)

plt.hist(x, bins=20)
plt.show()
```

```python
x = np.random.rand(64,64)

plt.imshow(x, cmap='hot')
plt.colorbar()
plt.show()
```
